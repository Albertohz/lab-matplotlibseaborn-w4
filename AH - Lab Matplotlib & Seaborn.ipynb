#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

```python
# MEDIAN STEPS BY MONTH_ENCODED
# your code here

# MEDIAN STEPS BY WORK_OR_WEEKEND
# your code here

# for this plot and how to put together both of them I had to take examples from an answer given to a similar problem in stackoverflow.
```

```python
import matplotlib.pyplot as plt

# Bets please notice this is a folder on my own google collab environment
import os
if not os.path.exists('figures'):
    os.makedirs('figures')

# MEDIAN STEPS BY MONTH_ENCODED
plt.figure(figsize=(12, 4))

# median steps for each month
median_steps_by_month = fitbit.groupby('Months_encoded')['Steps'].median()

plt.bar(median_steps_by_month.index, median_steps_by_month.values, color='skyblue')
plt.title('Median Steps by Month')
plt.xlabel('Month (Encoded)')
plt.ylabel('Median Steps')
plt.xticks(median_steps_by_month.index, ['Jan', 'Feb', 'Mar', 'Apr', 'May', 'Jun', 'Jul', 'Aug', 'Sep', 'Oct', 'Nov', 'Dec'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('figures/median_steps_by_month.png')

# MEDIAN STEPS BY WORK_OR_WEEKEND
plt.figure(figsize=(12, 4))

# Group the data by 'Work_or_Weekend' and calculate the median steps for each category
median_steps_by_work_or_weekend = fitbit.groupby('Work_or_Weekend')['Steps'].median()

plt.bar(median_steps_by_work_or_weekend.index, median_steps_by_work_or_weekend.values, color='lightcoral')
plt.title('Median Steps by Workday/Weekend')
plt.xlabel('Workday or Weekend')
plt.ylabel('Median Steps')
plt.xticks([0, 1], ['Workday', 'Weekend'])
plt.grid(axis='y', linestyle='--', alpha=0.7)

plt.savefig('figures/median_steps_by_work_or_weekend.png')

plt.show()
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# your code here
# list of feature pairs
feature_pairs = [
    ('Minutes Lightly Active', 'Steps'),
    ('Minutes Very Active', 'Steps'),
    ('Minutes Sedentary', 'Steps')
]

# as before, folder to save the plots
import os
if not os.path.exists('figures'):
    os.makedirs('figures')

# Looping
for idx, (x_feature, y_feature) in enumerate(feature_pairs):
    plt.figure(figsize=(8, 6))
    plt.scatter(fitbit[x_feature], fitbit[y_feature], alpha=0.5, color='blue')
    plt.title(f'Scatter Plot: {x_feature} vs {y_feature}')
    plt.xlabel(x_feature)
    plt.ylabel(y_feature)
    plt.grid(linestyle='--', alpha=0.7)
    plt.savefig(f'figures/scatter_plot_{idx + 1}.png')

plt.show()
```
